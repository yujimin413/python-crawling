```python
# 템플릿 형태로 사용할 것이므로 외우거나 이해하지 않아도 괜찮음
import requests
from bs4 import BeautifulSoup

response = requests.get("https://startcoding.pythonanywhere.com/basic")
html = response.text
soup = BeautifulSoup(html, 'html.parser') # 문자열 형태의 html을 html.parser가 태그 객체로 하나씩 잘라서 soup에 저장
 
# 이제 아래처럼 태그를 넘겨주면 됨. 이를 위해 선택자를 만드는 과정이 중요!
# soup.select_one("선택자")
soup.select_one(".brand-name")
soup.select_one(".brand-name").text
soup.select_one(".brand-name").attrs
# href 부분만 출력하고 싶을 때 사용.
# 딕셔너리 형태의 데이터 출력.
# 딕셔너리의 키 값에 접근해서 href 정보만 가져올 수도 있음
soup.select_one(".brand-name").attrs['href']
soup.select_one(".product-name > a").text
soup.select_one(".product-name > a").attrs['href']
soup.select_one(".product-price").text
# 출력 결과의 \n \t 문자 제거하려면? .strip() 추가
# strip(): 앞뒤 공백 제거
soup.select_one(".product-price").text.strip() 
#replace('변경 전 문자', '변경 후 문자') : 문자열 교체
soup.select_one(".product-price").text.strip().replace(',', '').replace('원', '')
```
